## 4. Modelo de regresion de la variables positive y negative

```python
pip install statsmodels 
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
reg=sm.OLS.from_formula('positive~negative',
                        data=df)
resolve=reg.fit()
resolve.summary()
```

### El valor de Durbin-Watson es de 0.004 indicando que existe multicolinealidad y los datos tienen una correlacion muy fuerte. Podemos notar que tienen una relacion linear al ver la R

## 5. Pivot Table pie chat

```python
pivot = pd.pivot_table(df,
                       index = 'states',
                       values = 'positive', 
                       aggfunc = 'count',
                      )
pivot
```

```python
table = df.groupby(['states']).size()
table = table.to_frame()
table
```

```python
table = table.rename(columns = {0:'pivot'})
table = table.reset_index()
table
```

```python
table['pivot'].plot(kind = 'pie',
                  autopct = '%1.1f%%',
                  startangle = 90,
                  shadow = True,
                  labels = None,
                  pctdistance = 1.2,
                 )
plt.legend(labels = table.states)
plt.title('Porcentaje de casos positivos por estado')
```

### El 83.1%de los casos positivos fueron en el estado 56.
